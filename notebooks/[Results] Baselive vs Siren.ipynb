```python
import pandas as pd
from pathlib import Path
import yaml
import json
import pickle
import torch
import flatdict
import matplotlib.pyplot as plt
```

```python
path_to_linear = '../exp/baseline_linear/linear_interpolation.pkl'

with open(path_to_linear, "rb") as f:
    res_lin = pickle.load(f)


path_to_siren = '../exp/tmp/best_model.pth'
res_sin = torch.load(path_to_siren)
```

```python
lin_rmse_phase_freq = res_lin["metrics"]["test"]["rmse_phase_freq"]
lin_rmse_phase = res_lin["metrics"]["test"]["rmse_phase"]
lin_rmse_time = res_lin["metrics"]["test"]["rmse_time"]
lin_lsd = res_lin["metrics"]["test"]["lsd"]
lin_lsd_freq = res_lin["metrics"]["test"]["lsd_freq"]

print(lin_rmse_phase)
print(lin_rmse_time)
print(lin_lsd)
```

```python
sin_rmse_phase_freq = res_sin["results"]["rmse_phase_freq"]
sin_rmse_phase = res_sin["results"]["rmse_phase"]
sin_rmse_time = res_sin["results"]["rmse_time"]
sin_lsd = res_sin["results"]["lsd"]
sin_lsd_freq = res_sin["results"]["lsd_freq"]

print(sin_rmse_phase)
print(sin_rmse_time)
print(sin_lsd)
```

```python
plt.plot(sin_rmse_phase_freq, label="SIREN")
plt.plot(lin_rmse_phase_freq, label="LINEAR")
plt.xlabel('Freq (Hz)')
plt.legend()
plt.title("Phase Error")
plt.show()
```

```python
plt.plot(sin_lsd_freq, label="SIREN")
plt.plot(lin_lsd_freq, label="Linear")
plt.xlabel('Freq (Hz)')
plt.legend()
plt.title("LSD Error")
plt.show()
```
